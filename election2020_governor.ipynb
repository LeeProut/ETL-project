### Governor's Race Completed Result

```python
#merging county and candidate dfs for all counties, candidates, and percentages reported
gov_mrgd_df = gov_cn_df.merge(gov_cand_df, how='left', on=['state', 'county'])
gov_mrgd_df
```

```python
gov_result_df = gov_mrgd_df.loc[gov_mrgd_df['percent'] >= 90]
gov_result_cleaned_df = gov_result_df.rename(columns={"votes": "candidate_votes",
                                                      "percent": "percent_count"}).copy()
gov_completed_result_df = gov_result_cleaned_df[['state', 'county',
                                                 'candidate', 'party', 
                                                 'candidate_votes', 'won', 
                                                 'percent_count']].copy()
gov_completed_result_df.head()
```

### Governor's Race Winner Result

#### By County

```python
gov_cn_winner_df = gov_completed_result_df.loc[gov_completed_result_df['won'] == True]
gov_cn_winner_df
```

#### By State

```python
gov_st_winner_votes = gov_completed_result_df.groupby(['state', 'candidate']).agg({'candidate_votes':sum}).reset_index(['state', 'candidate'])
gov_st_winner_df = gov_st_winner_votes[gov_st_winner_votes['candidate_votes'] == gov_st_winner_votes.groupby(['state'])['candidate_votes'].transform('max')]
gov_st_winner_df
```
